## Data Visualization

```python
def plot_distribution(df,title): 
    numerical_cols=df.select_dtypes(include=["number"]).columns 
    plt.figure(figsize=(12,8))
    for i,col in enumerate(numerical_cols[:6],1): 
        plt.subplot(2,3,i) 
        sns.histplot(df[col],kde=True,bins=30) 
        plt.title(col) 
    plt.suptitle(title) 
    plt.tight_layout() 
    plt.show() 
    
plot_distribution(price_data, "Price Data Distributions")
plot_distribution(client_data, "Client Data Distributions") 
```
