```python
import pandas as pd  
```

```python
clean_data=pd.read_csv("clean_data_after_eda.csv")  
clean_data
```

```python
price_data = pd.read_csv('price_data (1).csv')
```

```python
client_data = pd.read_csv("client_data (1).csv") 
```

```python
date_cols = ["date_activ", "date_end", "date_modif_prod", "date_renewal"]
for col in date_cols:
    clean_data[col] = pd.to_datetime(clean_data[col], errors='coerce')
```

```python
price_data["price_date"] = pd.to_datetime(price_data["price_date"], errors='coerce')
```

```python
# contract duration
clean_data["contract_duration_days"] = (clean_data["date_end"] - clean_data["date_activ"]).dt.days
```

```python
# time since last modification
clean_data["days_since_last_modif"] = (pd.to_datetime("today") - clean_data["date_modif_prod"]).dt.days
```

```python
# days until renewal
clean_data["days_until_renewal"] = (clean_data["date_renewal"] - pd.to_datetime("today")).dt.days
```

```python
# monthly consumption trend
clean_data["consumption_trend"] = clean_data["cons_last_month"] - (clean_data["cons_12m"] / 12) 
```

```python
# 6-month and yearly price variability
clean_data["yearly_price_variability"] = clean_data["var_year_price_peak"] - clean_data["var_year_price_off_peak"]
clean_data["6m_price_variability"] = clean_data["var_6m_price_peak"] - clean_data["var_6m_price_off_peak"]
```

```python
merged_data = clean_data.merge(price_data, on="id", how="left")
```

```python
final_data = merged_data.merge(client_data, on="id", how="left")
```
